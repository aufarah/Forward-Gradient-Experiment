# So what's that

Slope they said, mathematically when we said wa want to find gradient of F at point X, we find its partial derivation at X. Geometrically the gradient at X is the biggest of tangent vectors that touch F(X). 

## So what's the issue with Estimation?
For single parameter (only x here), finding gradient of F at X is easy since there's only one tangent there:


But given there're more than one (even billion in ML), we can find a bunch of tangents exist on tangent plane, we need to find the biggest one (or some dude said the average):


## Okay but why we need Estimation?
Gradient descent is the basic approach on NN training, and the popular way to get the gradient is by Backpropagation (Analytically find the derivation equation and voila). The backprop worked well if we know everything (ie stated as math equations) from inputs to objective function (mostly cost);  

But on some reinforced/unsupervised stuff like game training the relation between action (ie go down) and the score (+100) is totally unknown and kinda stupid to make another prediction.

# Estimating simple single parameter

```python
import numpy as np

#this is function
def quadrate(x):
    return x**2

#this is actual grad
def grad_quad(x):
    return 2*x

#this is estimation grad
def est_grad(x,func):
    h = 0.0001
    grad = func(x+h) - func(x)
    grad /= h
    return grad

grad_quad(1), est_grad(1,quadrate)
```

# Estimating 2 params
Here the algorithm:
1. Choose vector(s) v
1. Find partial derivation (tangents)
1. Print all tangents
1. Choose the average

```python
import numpy as np
#let's packed all of them, suppose the point is defined as np array
# the quadrate is np array dot itself

#this is function
def quadrate(point):
    # x**2 + y**2
    o = point*point
    return o.sum(axis=1)

#this is actual grad
def grad_quad(point):
    # 2*x, 2*y
    return 2*point

#this is estimation grad
def est_grad(xs,func,average=True):
    h = 0.0001
    v = np.random.uniform(1e-10,1,[1,xs.shape[1]])
    grads = []
    for x in xs:
        grads_sample = []
        for v_i in v:
            di = x+h*v_i
            each_param_h = np.identity(di.shape[0]) * di
            each_param   = np.identity(x.shape[0]) * x
            gr = func(each_param_h) - func(each_param )
            gr = gr/(h*v_i)
            grads_sample.append(gr)
        grads_sample = np.array(grads_sample)
        grads_sample = np.average(grads_sample,axis=0)
        grads.append(grads_sample)
    grads = np.array(grads) 
    if (average):
        return np.average(grads,axis=0)
    return grads #np.average(grads,axis=0)

point = np.array([[2,0]])
print("Gradien aseli: ",grad_quad(point))
print("Gradien nebak: ",est_grad(point,quadrate,False))
```

## 0 hidden layer
Usual work to predict `x1 and x2`, we use 2 input and 1 output hence 3 parameters (2w 1b)

The algorithm is as usual with normal training with backprop, the differences are on finding the gradient:  
We have :
 - several set of x-y pairs (here 4), 
 - forward_all method, forwarding all pairs,
 - cost_all method, get cost of all pairs
The Algorithm:
1. Setup h and v
1. Save initial params (the 3 parameters)
1. Set blank grad_set
2. For v_i in v:
    1. Set final params
    1. Set blank total gradient
    1. Do partial derivation for every i,param in params:
        1. Find cost with original param, store as f_param*
        2. Update param partially (just the `params[i]`)
        3. Find cost with updated param, store as f_param_h
        4. Store f_param_h - f_param as sels
        5. Get partial gradient by dividing sels with `h*v_i[i]`
        6. Stack partial gradient on total gradient
    1. Stack total gradient on grad_set
3. Find averaged grad_set
        
`*`. Some how we can't store f_param before all loop, it ruins training
